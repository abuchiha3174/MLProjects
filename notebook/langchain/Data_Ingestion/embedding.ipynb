```python
from langchain_openai import OpenAIEmbeddings
embeddings_3 = OpenAIEmbeddings(model = "text-embedding-3-large", dimensions = 1024)
embeddings_3
```

```python
text = "This is an example text."
query_vector = embeddings_3.embed_query(text)
query_vector
```

```python
## Text Loader - To load any txt file.
from langchain_community.document_loaders import TextLoader

loader = TextLoader('speech.txt')
loader
## will help in create documents
txt_documents = loader.load()
txt_documents
```

```python
speech = ""
with open("speech.txt") as f:
    speech = f.read()

speech
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
txt_splitter = RecursiveCharacterTextSplitter(chunk_size = 100, chunk_overlap = 20)
txt_doc = txt_splitter.split_documents(txt_documents)
txt_doc
```

```python
## Vector Embedding and Vector storeDB
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(txt_doc, embeddings_3)
```

```python
search_query = "government through all these bitter months because of that friends"
retrieved_results = db.similarity_search(search_query)
retrieved_results
```
